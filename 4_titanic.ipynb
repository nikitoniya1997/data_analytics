# Анализ данных по Титанику

```python
import pandas as pd
```

```python
# загрузка данных
data_0 = pd.read_csv("data/titanic.csv")
data_0.head(10)
```

```python
data_0.shape
```

```python
# удаление ненужных столбцов
data_1 = data_0.drop(['sibsp', 'parch','embarked', 'class', 'adult_male', 'deck', 'alive', 'alone'], axis=1)
data_1.head(10)
```

```python
data_1.info()
```

```python
# удаление строк с пустыми значениями
data_1 = data_1.dropna()
```

```python
# проверка на дубликаты
data_1.duplicated().sum()
```

```python
# удаление дубликатов
data_1 = data_1.drop_duplicates()
data_1.shape
```

```python
# сколько выживших в таблице
num_surv = data_1['survived'].sum()
num_surv
```

```python
# процент выживаемости
num_surv / data_1.shape[0] * 100
```

```python
# минимальный, средний и максимальный возрасты
data_1['age'].min(), data_1['age'].mean(), data_1['age'].max()
```

```python
# значение столбца 'survived' при условии 'age' == 0.42
mask = data_1['age'] == 0.42
data_1.loc[mask, 'survived']
```

```python
data_1.loc[mask, :]
```

```python
# узнаем данные по детям от 0.42 до 1 года
mask = (data_1['age'] >= 0.42) & (data_1['age'] <= 1)
data_1.loc[mask, :]
```

```python
# уникальные значения по столбцу 'embark_town'
data_1['embark_town'].unique()
```

```python
# группировка данных
data_1.groupby("survived")[['age', 'pclass']].mean()
```

```python
data_1.groupby(["survived", 'pclass'])[['age', 'fare']].mean()
```

```python
data_1.groupby(["survived", "who"])[['age', 'pclass']].mean()
```

```python
data_1.groupby(["survived", "who"])[['age']].count()
```

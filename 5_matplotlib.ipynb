```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Линейные графики

# 1-мерный массив
x_arr = np.arange(-10, 10)
x_arr
```

```python
y_arr = x_arr ** 3
y_arr
```

```python
plt.figure(figsize=(10,7))
plt.plot(x_arr, y_arr)
plt.show
```

```python
# совмещение нескольких графиков в одном окне
y_arr_1 = x_arr ** 2
y_arr_2 = x_arr ** 3

plt.plot(x_arr, y_arr_1)
plt.plot(x_arr, y_arr_2)

plt.show()
```

```python
# 1 - мерный массив из 100 значений
# в диапазоне от -10 до 10

x_arr_1 = np.linspace(-10, 10, 100)
x_arr_1
```

```python
# функция синуса
y_1 = np.sin(x_arr_1)

# функция гиперболического тангенса
y_2 = np.tanh(x_arr_1)

# фукнция сигмоиды
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y_3 = sigmoid(x_arr_1)

plt.plot(x_arr_1, y_1)
plt.plot(x_arr_1, y_2)
plt.plot(x_arr_1, y_3)

plt.show()
```
